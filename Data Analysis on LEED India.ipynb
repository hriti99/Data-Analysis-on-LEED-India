**First importing the Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**LOAD THE CLEAN DATASET**

```python
data = pd.read_csv("cleaned_buildingdataset.csv")
```

```python
data
```

**SET THE STYLE FOR SEABORN PLOTS**

```python
sns.set_style("whitegrid")
```

**PLOT FOR DIFFERENT TYPE OF BUILDINGS**

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Title2', palette='Set2')  # Using Set2 color palette
plt.title('Count of Different Types of Buildings')
plt.xlabel('Building Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**DISTRIBUTION PLOT FOR BUILDING SIZES (SQUARE FOOTAGE)**


```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Field8', bins=20, kde=True, color='skyblue')  # Using skyblue color
plt.title('Distribution of Building Sizes')
plt.xlabel('Square Footage')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

**PLOT FOR CERTIFICATION LEVELS**

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='jvisibleproject', palette='Paired')  # Using Paired color palette
plt.title('Certification Status of Buildings')
plt.xlabel('Certification Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**PLOT FOR YEAR OF CONSTRUCTION**

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Field9', order=data['Field9'].value_counts().index, palette='Pastel1')  # Using Pastel1 color palette
plt.title('Count of Buildings by Year of Construction')
plt.xlabel('Year of Construction')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**PLOT FOR LOCATIONS**

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=data, x='Like', order=data['Like'].value_counts().index, palette='Set3')  # Using Set3 color palette
plt.title('Count of Buildings by Location')
plt.xlabel('Count')
plt.ylabel('Location')
plt.xticks(rotation=45)
plt.show()

```

```python
gold_platinum_buildings = data[(data['jvisibleproject'] == 'GOLD') | (data['jvisibleproject'] == 'PLATINUM')]
```

```python
print("Summary statistics for relevant columns:")
print(gold_platinum_buildings.describe())
```

**ADD ATTRIBUTES COLUMN IN THE DATASET**

```python


# Define a dictionary mapping certification levels to attributes
certification_attributes = {
    'GOLD': '60% Sustainable',
    'PLATINUM': '40% Sustainable',
    'REGISTERED': 'Less than 40% Sustainable'
    # Add more mappings as needed
}

# Create the new column "attributes" based on the values in the "jvisibleproject" column
data['attributes'] = data['jvisibleproject'].map(certification_attributes)

# Display the updated dataset with the new column
print(data.head())  # Display the first few rows to verify the changes


```

**Attributes will differentiate between Gold, Platinum, Registered**

```python
data.to_csv("updated_dataset.csv", index=False)  # Replace "updated_dataset.csv" with the desired file name
```
